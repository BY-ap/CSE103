## Hash Functions
We want a set of hash functions $h_1,\ldots, h_k$, that map arbitrary numbers to a pseudo random number in the range $0,\ldots,m-1$

To compute $h_i(j)$ we combine the index of the hash function $i$ with the number to be hashed value $j$, to create a seed for the random number generator. We use the pseudo-random number generated by `random.randrange` output by the generator as $h_i(j)$.

```python
class Hash:
    def __init__(self,m):
        random.seed() # set the seed of the hash function using the system timer
        self.seed=random.randrange(1000000)
        self.m=m
    def Map(self,i):
        """ 
        map i into a random number in the range 0..m-1
        """
        combined=self.seed+i
        random.seed(combined)
        return random.randrange(self.m)
```

Demo of a hash function. Note that $h_1(j)$ always maps to the same number, but the hash function is different each time a Hash object is created

```python
h=Hash(100)
for i in range(10):
    j=i%5
    print('(',j,h.Map(j),end='), ')
print()
```

## Bloom filters

```python
class Bloom:
    def __init__(self,k,m):
        self.k = k
        H=[]   # initialize a list of k Hash functions
        for i in range(k):
            H.append(Hash(m))
        self.H=H
        self.m = m
        self.T = np.zeros(m)
    
    def check(self,j):
        """Check whether all k entries are 1"""
        #print(sum(self.T))
        for i in range(self.k):
            #print(j,self.H[i].Map(j),self.T[self.H[i].Map(j)])
            if self.T[self.H[i].Map(j)]==0:
                return False
        return True

    def mark(self,j):
        """Check whether all k entries are 1"""
        for i in range(self.k):
            self.T[self.H[i].Map(j)]=1
            
    def level(self):
        return sum(self.T)/self.m
            
    def reset(self):
        self.T=np.zeros(self.m)
```

```python
# insert elements into the hash table until an element is marked before it is inserted (false positive)
def failure_probability(m,n,k=5,tests=10,pokes=100):
    """ put n elements in a bloom filter of order k and check for failure probability"""
    failures=0
    for test in range(tests):
        b=Bloom(k,m)
        for i in range(n):
            b.mark(i)
        #print(n,m,b.level())
        for poke in range(pokes):
            i+=1
            if b.check(i):
                failures+=1
    return failures/(tests*pokes)
```

```python
n=1000
failures={}
k_values=[1,2,6,9,40]
for k in k_values:
    print('k=',k,end=', ')
    for m in range(3*n,20*n,2*n):
        f=failure_probability(m,n,k=k)
        failures[(k,m)]=f
```

```python
ms=np.array([m for (k,m) in failures if k==1])
ms
```

```python
def bound(k,n,m):
    return (1-exp(-(k*n)/m))**k
```

```python
figure(figsize=[15,8])
for k in k_values:
    plot(ms/n, [failures[(k,m)] for m in ms],label='estim k=%d'%k)
    plot(ms/n, [bound(k,n,m) for m in ms],'--',label='bound k=%d'%k)
legend()
grid()
xlabel('m/n')
ylabel('probability of false positive')
title('Bloom performance for n=%d and different m,k'%(n));
```
